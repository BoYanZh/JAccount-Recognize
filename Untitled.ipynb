```python
def get_image():
    counts = 0    
    for file in glob.glob("binary_img\*.bmp"):
        filename = 'out_img\\' + file.split('.')[0].split("\\")[1] + '.bmp'
        #print(file)
        im = cv2.imread(file,0)
        #print(im.shape[1],im.shape[0])
        im = 255 - clean_img(255 - im)
        #plt.imshow(im, cmap='Greys_r')
        cv2.imwrite(filename,im)
        counts += 1
    print(counts)
```

```python
def get_dynamic_binary_image():
    count = 0
    for file in glob.glob("captcha\*.jpg"):
        filename = 'binary_img/' + file.split('.')[0].split("\\")[1] + '.bmp'
        im = cv2.imread(file)
        im = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY) #灰值化
        # 二值化
        #th1 = cv2.adaptiveThreshold(im, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 21, 1)
        ret,th1 = cv2.threshold(im,180,255,cv2.THRESH_BINARY)
        cv2.imwrite(filename,th1)
        count += 1
    print(count)
```

```python
def binary_img(im):
    im = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
    ret,th1 = cv2.threshold(im,180,255,cv2.THRESH_BINARY)
    return th1
```

```python
def clean_img(im):
    try:
        #print(im.shape[1],im.shape[0])
        while True:
            if im[:,0].sum() == 0:
                im = np.delete(im, 0, axis=1)
            else:
                break
        while True:
            if im[:,-1].sum() == 0:
                im = np.delete(im, -1, axis=1)
            else:
                break
        while True:
            if im[0].sum() == 0:
                im = np.delete(im, 0, axis=0)
            else:
                break
        while True:
            if im[-1].sum() == 0:
                im = np.delete(im, -1, axis=0)
            else:
                break
        return im
    except:
        return []
```

```python
def split_img(im):
    imgList = []
    im = 255 - im
    cutList = [0]
    index = 0
    while index < im.shape[1] and len(cutList) <= 5:
        tmpSum = im[:,index].sum()
        if tmpSum == 0:
            cutList.append(index)
            tmpSum = im[:,index + 1].sum()
            while tmpSum == 0:
                im = np.delete(im, index + 1, axis=1)
                tmpSum = im[:,index + 1].sum()
        index += 1
    cutList.append(im.shape[1])
    if len(cutList) < 4:
        return []
    elif len(cutList) < 6:
        while True:
            maxWid = 20
            for i in range(len(cutList) - 1):
                if cutList[i + 1] - cutList[i] > maxWid:
                    maxWid = cutList[i + 1] - cutList[i]
                    md = int((cutList[i]+cutList[i + 1])/2)
            if maxWid > 20:
                bias = 4
                minSum = 255 * im.shape[0]
                for i in range(md - bias, md + bias + 1):
                    tmpSum = im[:,i].sum()
                    if tmpSum <= minSum:
                        tmpCut = i
                        minSum = tmpSum
                cutList.append(tmpCut)
                cutList.sort()
            else:
                break
    for i in range(len(cutList) - 1):
        img = im[:,cutList[i]:cutList[i+1]]
        img = clean_img(img)
        if len(img) != 0 and img.sum() != 0:
            img = cv2.resize(img,(28,28))
            imgList.append(img)
    return np.array(imgList)
```

```python
x_train = np.load("img.npy")
y_train = np.load("letter.npy")
x_train = x_train/255
print(x_train.shape)
print(y_train.shape)
from keras.utils import to_categorical
y_train = to_categorical(y_train)
plt.imshow(x_train[10], cmap='gray')
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
from keras.callbacks import ModelCheckpoint

model = Sequential()

model.add(Dense(512, input_shape=(784,), kernel_initializer='he_normal'))
model.add(Activation('relu'))
model.add(Dropout(0.2)) 

model.add(Dense(512, kernel_initializer='he_normal'))
model.add(Activation('relu'))
model.add(Dropout(0.2)) 

model.add(Dense(26))
model.add(Activation('softmax'))

filepath="weights.best.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='val_acc', verbose=1, save_best_only=True,
mode='max')
callbacks_list = [checkpoint]

model.compile(loss='categorical_crossentropy',optimizer='sgd', metrics=['accuracy'])

model.summary()
```

```python
x_train = x_train.reshape(x_train.shape[0],784)
model.fit(x_train, y_train, epochs=8, validation_split=0.2, callbacks=callbacks_list)
```

```python
score = model.evaluate(x_train, y_train, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
